```python
# Mapping definitions 


def mapping(df,to_map):

    if to_map == "person_id":
        return subject_id_mapping(df)
    if to_map == "gender_concept_id":
        return  gender_concept_id_mapping(df)
    if to_map == "year_of_birth":
        return year_of_birth_mapping(df)
    if to_map == "race_concept_id":
        return race_concept_id_mapping(df)
    if to_map == "ethnicity_concept_id":
        return ethnicity_concept_id_mapping(df)
    else:
        raise ValueError(f"Error: {to_map} mapping not found")


def subject_id_mapping(person_df):

    person_id = person_df["subject_id"]

    return person_id


def gender_concept_id_mapping(person_df):

    gender_concept_id = 0 

    return gender_concept_id


def year_of_birth_mapping(person_df):

    #year_of_birth = person_df["dob"]
    year_of_birth = pd.to_datetime(person_df["dob"]).dt.year

    return year_of_birth

def race_concept_id_mapping(person_df):
    race_concept_id = 0 

    return race_concept_id


def ethnicity_concept_id_mapping(person_df):
    ethnicity_concept_id = 0

    return ethnicity_concept_id




```

```python
def transform_PATIENTS(PATIENTS_df, person_df):

    person_df["person_id"] = mapping(df = PATIENTS_df, to_map = "person_id")
    person_df["gender_concept_id"] = mapping(df = PATIENTS_df, to_map = "gender_concept_id")
    person_df["year_of_birth"] = mapping(df = PATIENTS_df, to_map = "year_of_birth")
    person_df["race_concept_id"] = mapping(df = PATIENTS_df, to_map = "race_concept_id")
    person_df["ethnicity_concept_id"] = mapping(df = PATIENTS_df, to_map ="ethnicity_concept_id")
    

    return(person_df)
    

    

transform_PATIENTS(PATIENTS_df,person_base_necessary)


```
